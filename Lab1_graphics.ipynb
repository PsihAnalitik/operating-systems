```python
import matplotlib.pyplot as plt
import pandas as pd

with open('Analyze_time.txt','r') as file:
    content = file.read()
    
column_names = ['Размер блока', 'Время обмена', 'Скорость Кбайт/с']
data = pd.read_csv('Analyze_time.txt', sep = ' ', header = None, names = column_names, index_col = False)
data 

data_sorted = data.sort_values(by='Размер блока', ascending=True)
plt.plot(data_sorted['Размер блока'],data_sorted['Скорость Кбайт/с'] )
plt.grid()
plt.xlabel(column_names[0])
plt.ylabel(column_names[2])

data_sorted
```
